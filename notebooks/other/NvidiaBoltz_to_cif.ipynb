This is a small script that

*   takes in a json file that Boltz implementation on [Nvidia](https://build.nvidia.com/mit/boltz2) servers produces

*   and outputs a cif file that you can download




The resulting file will be downloaded automatically. If this doesn't work, navigate to the left panel, open files (a folder icon), hover over a file name, click on three dots icon and click "Download".

```python
import json
from google.colab import files
```

```python
upload_dict = files.upload()
output_name='my_file' #@param {type:"string"}
upload_string = upload_dict[list(upload_dict.keys())[0]]
with open('task.json',"wb") as out: out.write(upload_string)
J = json.load(open('./task.json', 'r'))
cif = J['structures'][0]['structure']
with open(f'{output_name}.cif', 'w') as fh:
  fh.write(cif)
print(f"Wrote a cif file named {output_name}.cif")

files.download(f"{output_name}.cif")
```
